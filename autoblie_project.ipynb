```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\punit\Downloads\automobile.csv')
df.info()
df.head()
```

```python
df=df.drop_duplicates().reset_index(drop=True)
df.head()
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['symboling'],inplace= True)
```

```python
df.info()
```

```python
missing_door=df[df['num-of-doors'].isnull()]
print(missing_door)
```

```python
df['num-of-doors']=df['num-of-doors'].fillna('four')
```

```python
df.drop(columns=['fuel-type','normalized-losses','peak-rpm','bore','stroke','fuel-system','compression-ratio'],inplace=True)
```

```python
df.info()
```

```python
#To check the distribution of cars across different body style
```

```python


sns.countplot(data=df,y='body-style',color='gold')
plt.grid(linestyle=':')
plt.title('numers of cars per bodytype')
plt.title('Numbers of cars per body type')
```

```python
#if we want to data in a high to low order
plt.Figure(figsize=(6,4))
sns.countplot(data=df,y='body-style',order=(df['body-style'].value_counts().index))
```

```python
#To check the distribution of cars acrcoss different body style with include doors numbers
```

```python
plt.Figure(figsize=(6,4))
sns.countplot(data=df,y='body-style',hue='num-of-doors')
plt.title('Numbers of car per bodytype and door numbers',y=1.02)
```

```python
# to check the average engine size in each bodystyle
plt.Figure(figsize=(6,4))
sns.barplot(data=df,x='engine-size',y='body-style',estimator='mean',errorbar=None)
plt.grid(linestyle=':')
plt.title('Average engine size in each bodystyle')
```

```python
#how does the engine size vary across different body style splits over number of doors
```

```python
#barplot
sns.barplot(data=df,x='engine-size',y='body-style',hue='num-of-doors',estimator='mean',errorbar=None)
plt.title('Average engine size vary across different body style splits over number of doors')
```

```python
#catplot
barplot=sns.catplot(kind='bar',data=df,y='body-style',x='engine-size', col='num-of-doors',estimator='mean',errorbar=None)
barplot.set_axis_labels('Average engine size','Car bodystyle')
barplot.fig.suptitle('Average engine size vary across different body style splits over number of doors',y=1.08)
```

```python
#by use of pivot table
pivot=df.pivot_table(index='body-style',columns='num-of-doors', values='engine-size',aggfunc='mean')
pivot.plot(kind='barh', stacked=False)
```

```python
#what is the distribution of car price
```

```python
sns.histplot(data=df,x='price',bins=10,color='linen')
plt.ylabel('number of cars')
plt.title('distribution of price' )
```

```python
#what is the distribution of price of cars split over bodystyle
sns.histplot(data=df,x='price',hue='body-style', bins=10,multiple='stack')
plt.ylabel('number of cars')
plt.title('distribution of price split over bodystyle')
```

```python
# using histogram
sns.histplot(data=df,x='city-mpg',bins=10)
plt.axvline(19,c='r')
plt.axvline(24,c='r')
plt.axvline(30,c='r')
plt.title('distribution of city mileage of cars' )
```

```python
sns.boxplot(data=df,x='city-mpg',y='body-style',hue='num-of-doors')
plt.title('distribution of city mileage splits over car body style and number of doors')
```

```python
# what is the distribution of horse power across the cars
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(data=df,x='horsepower')
plt.title('distribution of horse power',fontsize=12)
plt.xlabel('Horsepower',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.subplot(1,2,2)
sns.kdeplot(data=df,x='horsepower',fill=True)
plt.title('Distribution of horse power')
plt.xlabel('horsepower',fontsize=12)
plt.ylabel('Probabilty Density',fontsize=12)
```

```python
sns.histplot(data=df,x='horsepower',kde=True)
plt.title('distribution of horsepower')
```
